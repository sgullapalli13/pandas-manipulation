```python
import pandas as pd
import numpy as np
```

```python
# Initialise data to lists. 
data = [{'name': 'Alice', 'group': 'Student', 'createddate':'2019-01-03'}] 
# Creates DataFrame. 
df = pd.DataFrame(data, index=range(50)) 
```

```python
df['createddate'] = np.random.choice(pd.date_range('2019-01-01', '2019-01-10'), len(df))

names = ['Alice', 'Bob', 'Charlie', 'David', 'Edith', 'Fred', 'Greg', 'Holly', 'Isabel', 'John', 'Kevin', 'Luna', 'Matt']
groups = ['group.Student','group.Faculty','group.Admin']
df['name'] = np.random.choice(list(names), len(df))
df['group'] = np.random.choice(list(groups), len(df))
```

```python
#total rows in df
len(df)
```

```python
#remove rows of weekends from the dataframe
_weekends=False
if not _weekends:
    df_weekday = df[df.createddate.dt.dayofweek<5]
df_weekday
```

```python
#get the counts of how many accesses per day
df_daily = df.groupby(pd.Grouper(key='createddate', freq = 'D')).size().reset_index().rename(columns={"createddate":"createddate", 0:"count"})
df_daily
```

```python
#reset
#unique names per day (does not count same person clicking on same day)
df_unique = df[~pd.to_datetime(df.createddate).groupby(df.name).diff().dt.days.lt(1)]
# df_unique.sort_values(by=['createddate','name'], inplace = True)
df_unique_sort = df_unique.sort_values(by=['createddate','name'])
df_unique_sort
```

```python
#to get counts of name appearance
df_unique_count = df_unique.groupby([df['name']]).size().to_frame(name = 'count').reset_index()
df_unique_count
```

```python
df_unique_count.sort_values(by='count', inplace = True)
df_unique_count
```

```python
#get counts of accesses grouped by role
app_df = []
app_dict = {}
num = 0
_app_list = df['group'].unique()

_daterange = "D"

for i in _app_list:
    temp_df = df[df['group']==i].groupby(pd.Grouper(key='createddate', freq = 'D')).size().reset_index().rename(columns={"createddate":"createddate", 0:"count"})
        #final_df
    app_df.append(temp_df)

    app_dict.update({i:app_df[num]})
    num+=1
    
group_arr=[]
for i in _app_list:
    string = i.split('.')[1] + ": " + str(app_dict[i]['count'].sum())
    group_arr.append(string)
    
group_arr
```
